## Chapter 7-1 데이터 불러오기 <a class="anchor" id="chapter7-1"></a>
1. 데이터 분석의 첫 단계는 데이터셋을 불러와 각 열의 의미와 값을 정확히 파악하는 것이다. 
    - 이 과정을 탐색적 데이터 분석이라고 한다.(EDA: Exploratory Data Analysis)
    - 데이터 구조 이해: 각 열이 어떤 데이터를 담고 있는지 파악한다.
    - 데이터 품질 확인: 결측치, 이상치, 중복값 여부를 확인한다.
    - 페턴 및 관계 탐색: 변수 간 관계를 파악하고, 데이터에서 숨겨진 패턴을 찾는다.
    - EDA를 소홀히 하는 경우 데이터의 특성을 제대로 이해하지 못해 분석이 왜곡되거나 잘못된 결론에 도달할 수 있다.

2. IMDb(Internet Movie Database)에 있는 상위, 1,000개 영화와 TV 쇼에 대한 데이터를 이 장에서 사용한다.
    - imdb_top_1000.csv

3. 캐글(Kaggle)은 데이터 과학과 머신러닝 분야에서 가장 유명한 온라인 플랫폼이다.
    - 자신이 소유한 데이터를 업로드해 다른 사람과 공유할 수 있다.

```python
# read_csv() 함수를 하용해 imdb_top_1000.csv 파일을 읽어온다.
import pandas as pd

# CSV 파일 경로
file_path = "./data/imdb_top_1000.csv"
# CSV 파일 불러오기
df = pd.read_csv(file_path)
```

### Chapter 7-2-2 기본 통계 정보 확인하기 <a class="anchor" id="chapter7-2-2"></a>
1. 데이터셋의 기본 통계 정보를 확인하는 것은 데이터 분석의 중요한 단계이다.
    - 데이터셋의 크기, 열의 이름과 데이터 타입, 결측치 여부 등을 파악할 수 있다.
    - 이를 통해 데이터의 품질을 평가하고, 이후 분석에서 어떤 전처리 작업이 필요한지 결정할 수 있다.

2. describe() 메서드를 사용해 기본 정보를 확인할 수 있다.
    - count: 각 열의 유효한 값의 개수
    - mean: 각 열의 평균값
    - std: 각 열의 표준편차
    - min: 각 열의 최솟값
    - 사분위 점수: 25%, 50%, 75%에 해당하는 값
    - max: 각 열의 최댓값 

3. 표준편차 해석방법
    - 표준편차는 데이터가 평균에서 얼마나 퍼져 있는지를 나타내는 지표이다.
    - 표준편차가 크면 데이터가 평균에서 멀리 퍼져 있고, 표준편차가 작으면 데이터가 평균에 가까이 모여 있음을 의미한다
        - 예를 들어, IMDB_Rating 열의 표준편차가 0.5라면, 대부분의 영화 평점이 평균에서 0.5점 이내에 분포하고 있음을 나타낸다.
    - 데이터의 분포 범위에 대한 이해를 돕는다.
        - 평균 +-1 표준편차 범위 내에 데이터의 약 68%가 포함되고, 평균 +-2 표준편차 범위 내에 약 95%가 포함된다는 점을 활용할 수 있다.

## Chapter 7-3 영화 평점과 참여주 수 분석하기 <a class="anchor" id="chapter7-3"></a>
### Chapter 7-3-1 영화 평점의 평균 구하기 <a class="anchor" id="chapter7-3-1"></a>


```python
# IMDB 영화 평점의 평균 계산
average_rating = df['IMDB_Rating'].mean()
print("평점 평균: ", average_rating)
```

### Chapter 7-3-2 최고 평점 영화 찾기 <a class="anchor" id="chapter7-3-2"></a>

```python
# IMDB 영화 중 최고 평점 구하기
max_rating = df['IMDB_Rating'].max()
print("최고 평점: ", max_rating)
```

```python
# 불리언 인덱싱(boolean indexing)을 통하여 최대평점 영화 찾기
highest_rated_movie = df[df['IMDB_Rating'] == max_rating]
print("최고 평점 영화의 제목과 평균:\n", highest_rated_movie[['Series_Title', 'IMDB_Rating']])
```

### Chapter 7-3-3 최저 평점 영화 찾기 <a class="anchor" id="chapter7-3-3"></a>

```python
# IMDB 영화 중 최저 평점 구하기
min_rating = df['IMDB_Rating'].min()
print("최저 평점: ", min_rating)
```

```python
# 불리언 인덱싱(boolean indexing)을 통하여 최저평점 영화 찾기
lowest_rated_movie = df[df["IMDB_Rating"] == min_rating]
print("최저 평점 영화의 제목과 평균:\n", lowest_rated_movie[['Series_Title', 'IMDB_Rating']])
```

```python
# 최저 평점을 받은 영화의 수
print("최저 평점을 받은 영화의 수: ", len(lowest_rated_movie))
```

### Chapter 7-3-4 전체 평가 참여자 수와 영화당 평균 참여자 수 구하기 <a class="anchor" id="chapter7-3-4"></a>

```python
# No_of_Votes 열의 합을 구한다.
total_votes = df['No_of_Votes'].sum()
print("총 투표 수: ", total_votes)
```

```python
# No_of_Votes 열의 평균 계산
average_votes = df['No_of_Votes'].mean()
print("평균 투표 수: ", average_votes)
```

## Chapter 7-4 셀프체크 <a class="anchor" id="chapter7-4"></a>

```python
# IMDB 데이터 중 참여자 수가 평균 이상인 영화면 필터링해 총 몇편인지 계산
# 참여자 평균 구하기
average_vote = df['No_of_Votes'].mean()
print("average_vote : ",average_vote)

over_average_vote_movis = df[df['No_of_Votes'] >= average_vote];
print(len(over_average_vote_movis))
```

```python
# IMDB 전체 영화의 평균 수익
# 결측치 계산하기
missing_values = df['Gross'].isnull().sum()

# 결측치가 있는 행 제거
df_without_missing = df.dropna(subset=['Gross'])

# "," 제거하하기
df_without_missing['Gross'] = df_without_missing['Gross'].str.replace(',','')

# Gross 열을 숫자형으로 변환하기
df_without_missing['Gross'] = df_without_missing['Gross'].astype(float)

# Gross 열의 평균 계산하기
average_gross = df_without_missing['Gross'].mean()
print("IMDB 전체 영화의 평균 수익: ", average_gross)
```

```python
# IMDB 전체 영화의 평균 수익
# "," 제거하하기
df['Gross'] = df['Gross'].str.replace(',','')

# 숫자로 변환 불가능한 값 NnN처리
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')

# 결측치를 제거하고 평균 구하기
average_gross = df['Gross'].dropna().mean()
print("IMDB 전체 영화의 평균 수익: ", average_gross)
```

```python
# IMDB 영화 중 수익이 가장 높은 영화의 제목과 수익 출력하기
# 가장 높은 수익 구하기
max_gross = df['Gross'].max()

# 수익이 가장높은 영화의 제목과 수익 출력
df_max_gross = df[df['Gross'] == max_gross ]
df_max_gross[['Series_Title', 'Gross']]
```
