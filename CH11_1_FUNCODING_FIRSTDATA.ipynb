```python
import pandas as pd
# CSV 파일 불러오기
df = pd.read_csv('./data/imdb_top_1000.csv')
df.head()
```

```python
# 각 열의 데이터 타입 확인
print("각 열의 데이터 타입:")
print(df.dtypes)
```

```python
# 연도를 숫자형으로 변환(결측치 처리)
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce') 
# 결측치 제거
df = df.dropna(subset=['Released_Year']) 
# 연도를 정수형으로 변환
df['Released_Year'] = df['Released_Year'].astype(int) 
```

```python
import plotly.express as px 
# 연도별 평균 평점 계산
average_rating_per_year = df.groupby('Released_Year')['IMDB_Rating'].mean().reset_index() 
fig = px.line(
    average_rating_per_year, 
    x='Released_Year', y='IMDB_Rating', 
    title='연도별 IMDb 평균 평점', 
    labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
) 
fig.show() 
```

```python
fig = px.line(
 average_rating_per_year, 
 x='Released_Year', y='IMDB_Rating', 
 title='연도별 평균 IMDb 평점(spline 적용)', 
 line_shape='spline',
 labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
)
fig.show()
```

```python
fig = px.line(
 average_rating_per_year, 
 x='Released_Year', y='IMDB_Rating', 
 title='연도별 평균 IMDb 평점(hv 적용)', 
 line_shape='hv',
 labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
)
fig.show()
```

```python
fig = px.line(
 average_rating_per_year, 
 x='Released_Year', y='IMDB_Rating', 
 title='연도별 평균 IMDb 평점(마커 추가)', 
 markers=True, 
 labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
)
fig.show()
```

```python
fig = px.line(
 average_rating_per_year, 
 x='Released_Year', y='IMDB_Rating', 
 title='연도별 평균 IMDb 평점 (색상 변경)', 
 color_discrete_sequence=['green'], 
 labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
)
fig.show()
```

```python
fig = px.line(
 average_rating_per_year, 
 x='Released_Year', y='IMDB_Rating', 
 title='연도별 평균 IMDb 평점(템플릿 적용)', 
 template='simple_white',
 labels={'Released_Year': '개봉 연도', 'IMDB_Rating': '평균 평점'}
)
fig.show()
```

```python
import plotly.express as px
# 상위 10개 장르 필터링
top_genres = df['Genre'].value_counts().nlargest(10).index 
df_top_genres = df[df['Genre'].isin(top_genres)] 
df_top_genres
```

```python
# 상자 그림 그리기
fig = px.box(
    df_top_genres, 
    x='Genre', y='IMDB_Rating', 
    title='상위 10개 장르별 IMDb 평점 분포'
) 
# 그래프 설정 조정
fig.update_layout(
    width=900, # 그래프의 너비 설정
    height=600, # 그래프의 높이 설정
    xaxis_title='장르', # X축 이름 설정
    yaxis_title='IMDb 평점', # Y축 이름 설정
    xaxis_tickangle=-45, # X축 틱 레이블의 기울기 조정
    xaxis=dict(tickfont=dict(size=14)) # X축 틱 레이블의 글꼴 크기 설정
)
fig.show()
```

```python
# 상위 10개 장르 필터링
top_genres = df['Genre'].value_counts().nlargest(10).index
df_top_genres = df[df['Genre'].isin(top_genres)]
# 상자 그림 그리기
fig = px.box(
    df_top_genres, 
    x='Genre',
    y='IMDB_Rating',
    title='상위 10개 장르별 IMDb 평점 분포',  
    height=600,                          # 그래프 높이 설정
    color='Genre',                       # 장르별 색상 구분
    points='all',                        # 모든 데이터 포인트 표시
    notched=True                         # 중앙값의 신뢰 구간을 노치로 표시
)
# 그래프 설정 조정
fig.update_layout(
    xaxis_title='장르',                  # X축 이름 설정
    yaxis_title='IMDb 평점',             # Y축 이름 설정
    xaxis_tickangle=-45,                 # X축 레이블 기울기 조정
    xaxis=dict(tickfont=dict(size=14))   # X축 레이블 글씨 크기 설정
)
fig.show()
```

```python
# Gross 열의 쉼표 제거 후 숫자형으로 변환
df['Gross'] = df['Gross'].replace({',': ''}, regex=True).astype(float)
# 필요한 열 선택 및 결측치 제거
correlation_data = df[['IMDB_Rating', 'No_of_Votes', 'Gross', 'Meta_score']].dropna()
```

```python
# 상관계수 계산
correlation_matrix = correlation_data.corr()
# 히트맵 그리기
fig = px.imshow(correlation_matrix,
                text_auto=True,
                title='IMDb 데이터의 수치형 변수 간 상관관계',
                color_continuous_scale='RdBu')
# 그래프 크기와 텍스트 크기 조정
fig.update_layout(
    width=800,                           # 그래프의 너비
    height=800,                          # 그래프의 높이 
    title_font_size=20,                  # 제목 글씨 크기 
    xaxis=dict(tickfont=dict(size=14)),  # X축 틱 레이블 글씨 크기 
    yaxis=dict(tickfont=dict(size=14))   # Y축 틱 레이블 글씨 크기 
)
fig.show()
```
