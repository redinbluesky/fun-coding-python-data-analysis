### Chapter 5-1-1 CSV 파일 불러오기 <a class="anchor" id="chapter5-1-1"></a>
1. CSV(Comma-Separated Value) 파일은 데이터를 쉼표로 구분하여 저장하는 텍스트 파일 형식이다.
    - 엑겦에서도 쉽게 열 수 있으며, CVS 파일의 행은 엑셀의 행에, CVS 파일의 열은 엑셀의 열에 대응된다.
2. Pandas에서는 `read_csv()` 함수를 사용하여 CSV 파일을 불러올 수 있다.
    - `read_csv()` 함수는 CSV 파일의 경로를 인자로 받아 데이터를 DataFrame으로 반환한다.
    - `read_csv()` 함수는 다양한 옵션을 제공하여 CSV 파일을 불러올 때 데이터의 형식이나 구분자 등을 지정할 수 있다

```python
# Pandas 라이브러리를 사용해 /data/example.csv 파일을 불러와 보자.
import pandas as pd

# CSV 파일 경로
file_path = './data/example.csv'

# CSV 파일 불러오기
df = pd.read_csv(file_path)

# 불러온 데이터 확인
print(df)
```

## Chapter 5-2 데이터 확인하기 <a class="anchor" id="chapter5-2"></a>
1. 데이터를 불러온 후에는 데이터의 구조와 규모를 빠르게 확인하는 것이 중요하다.

2. 엑셀에서는 파일을 직접 열어서 데이터를 확인할 수 있지만, Pandas에서는 다양한 함수를 사용하여 데이터의 구조와 내용을 빠르게 파악할 수 있다.

### Chapter 5-2-1 초반 데이터 확인하기:head() <a class="anchor" id="chapter5-2-1"></a>
1. `head()` 함수는 DataFrame의 처음 몇 행을 반환하는 함수로, 기본적으로 처음 5행을 보여준다.
    - `head(n)`과 같이 인자를 주면 처음 n행을 반환한다.
    - 데이터의 구조와 내용을 빠르게 파악할 때 유용하다.

2. Pandas에서 데이터를 확인하는 함수는 별칭인 pd를 사용하지 않고, 시리즈나 데이터프레임이 저장된 변수면을 앞에 붙여 사용한다.
    - 데이터 프레임이 data라는 변수에 저장되어 있다면, `data.head()`와 같이 사용한다.

```python
# example.csv 파일을 불러온 후, head() 함수를 사용하여 처음 5행을 확인해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 처음 5행 확인하기
#   - 파이썬 표준 출력 방식을 사용하여 출력
print(data.head())

# 처음 3행 확인하기
#   - 주피터 노트북의 랜더링 방식을 사용하여 출력
data.head(3)
```

### Chapter 5-2-2 후반 데이터 확인하기:tail() <a class="anchor" id="chapter5-2-2"></a>
1. `tail()` 함수는 DataFrame의 마지막 몇 행을 반환하는 함수로, 기본적으로 마지막 5행을 보여준다.
    - `tail(n)`과 같이 인자를 주면 마지막 n행을 반환한다.
    - 데이터의 구조와 내용을 빠르게 파악할 때 유용하다.

```python
# example.csv 파일을 불러온 후, tail() 함수를 사용하여 마지막 5행을 확인해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 마지막 5행 확인하기
print(data.tail())

# 마지막 3행 확인하기
print(data.tail(3))
```

### Chapter 5-2-3 통계 요약하기:describe() <a class="anchor" id="chapter5-2-3"></a>
1. `describe()` 함수는 DataFrame의 기본 통계 정보를 요약하여 보여주는 함수이다.
    - 수치형 데이터에 대해 평균, 표준편차, 최소값, 25% 백분위수, 50% 백분위수(중앙값), 75% 백분위수, 최대값 등의 통계 정보를 제공한다.
    - 수치형 데이터만 요약하며, 비수치형 데이터가 혼재되어 있으면, 수치형 데이터에 대한 통계 정보만 제공한다.
    - `describe(include='all')`와 같이 인자를 주면 모든 데이터 타입에 대한 통계 정보를 제공한다.

```python
# example.csv 파일을 불러온 후, describe() 함수를 사용하여 데이터의 기본 통계 정보를 요약해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 데이터의 기본 통계 정보 요약하기
print(data.describe())
```

2. 수치 데이터 열에 대한 `describe()` 함수의 출력은 다음과 같은 열로 구성된다.
    - count 열의 값은 열에서 결측값이 아닌 데이터의 개수를 나타낸다.
        - 이 값이 데이터 프레임의 행 수와 일치하지 않는다면, 해당 열에 결측값이 존재한다는 것을 의미한다.
        - 결측값: 데이터가 존재하지 않는 경우를 나타내며, Pandas에서는 NaN(Not a Number)으로 표현된다.
    -   mean 열의 값은 열의 평균값을 나타낸다.
        - 평균값은 데이터의 중심 경향성을 나타내며, 데이터가 정규 분포를 따르는 경우에 유용하다.
        - 평균값이 극단적인 값에 영향을 받을 수 있으므로, 데이터의 분포를 함께 확인하는 것이 중요하다.
    - std 열의 값은 열의 표준편차를 나타낸다.
        - 표준편차는 데이터의 분산 정도를 나타내며, 값이 클수록 데이터가 평균값에서 멀리 퍼져 있음을 의미한다.
        - 표준편차가 0에 가까운 경우, 데이터가 평균값에 매우 밀집되어 있음을 나타낸다.
    - min 열의 값은 열의 최소값을 나타낸다.
        - 최소값은 데이터에서 가장 작은 값을 나타내며, 데이터의 범위를 이해하는 데 도움이 된다.
        - 최소값이 극단적인 값인 경우, 데이터에 이상치가 존재할 수 있으므로, 데이터의 분포를 함께 확인하는 것이 중요하다.
    - 25% 열의 값은 열의 25% 백분위수를 나타낸다.
        - 25% 백분위수는 데이터의 하위 25%에 해당하는 값을 나타내며, 데이터의 분포를 이해하는 데 도움이 된다.
        - 25% 백분위수와 75% 백분위수의 차이를 사분위 범위(IQR)라고 하며, 데이터의 분산 정도를 나타내는 지표로 사용된다.
    - 50% 열의 값은 열의 50% 백분위수(중앙값)를 나타낸다.
        - 중앙값은 데이터의 중심값을 나타내며, 데이터가 정규 분포를 따르지 않는 경우에도 유용한 중심 경향성 지표이다.
        - 중앙값이 평균값과 크게 차이가 나는 경우, 데이터에 이상치가 존재할 수 있으므로, 데이터의 분포를 함께 확인하는 것이 중요하다.
    - 75% 열의 값은 열의 75% 백분위수를 나타낸다.
        - 75% 백분위수는 데이터의 상위 25%에 해당하는 값을 나타내며, 데이터의 분포를 이해하는 데 도움이 된다.
        - 25% 백분위수와 75% 백분위수의 차이를 사분위 범위(IQR)라고 하며, 데이터의 분산 정도를 나타내는 지표로 사용된다.

```python
# example.csv 파일을 불러온 후, describe(include='all') 함수를 사용하여 모든 데이터 타입에 대한 통계 정보를 요약해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 모든 데이터 타입에 대한 통계 정보 요약하기
print(data.describe(include='all'))
```

3. 비수치에 대한 `describe()` 함수의 출력은 다음과 같은 열로 구성된다.
    - unique 열의 값은 열에서 고유한 값의 개수를 나타낸다.
        - 고유한 값의 개수가 많을수록, 해당 열이 범주형 데이터일 가능성이 높다.

        - 가장 빈번하게 나타나는 값이 전체 데이터에서 차지하는 비율이 높다면, 해당 열이 범주형 데이터일 가능성이 높다.
        - 값이 여러개인 경우 임의 하나가 top으로 나타난다.
    - freq 최빈값이 나타나는 빈도 수를 나타낸다.
        - 최빈값이 전체 데이터에서 차지하는 비율이 높다면, 해당 열이 범주형 데이터일 가능성이 높다.

4. 엑셀에서 수식을 사용해 통계를 구하는 것과 비슷하지만, `describe()` 함수는 여러 통계 정보를 한 번에 요약하여 제공하므로, 데이터의 전반적인 분포와 특성을 빠르게 파악할 수 있다.

## Chapter 5-3 개별 통계치 구하기 <a class="anchor" id="chapter5-3"></a>
1. `describe()` 함수는 데이터의 기본 통계 정보를 요약하여 제공하지만, 개별 통계치를 구할 때는 각 통계치에 해당하는 함수를 사용한다.
    - 예를 들어, 평균값을 구할 때는 `mean()` 함수를 사용하고, 중앙값을 구할 때는 `median()` 함수를 사용한다.
    - 개별 통계치를 구할 때는 데이터의 분포와 특성을 더 자세히 파악할 수 있다.

### Chapter 5-3-1 평균 구하기:mean() <a class="anchor" id="chapter5-3-1"></a>
1. `mean()` 함수는 DataFrame의 수치형 열에 대한 평균값을 계산하는 함수이다.
    - `mean()` 함수는 기본적으로 수치형 열에 대해서만 계산하며, 비수치형 열이 혼재되어 있으면, 수치형 열에 대한 평균값만 반환한다.
    - `mean()` 함수는 `skipna=True` 옵션을 사용하여 결측값을 무시하고 평균값을 계산할 수 있다.


```python
# example.csv 파일에서 'Age'와 'Salary' 열의 평균값을 구해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 'Age'와 'Salary' 열의 평균값 구하기
age_mean = data['Age'].mean()
salary_mean = data['Salary'].mean()

# 여러 열의 평균 한 번에 구하기
#   - 시리즈 형태로 반환: 열이름은 인덱스, 평균값은 값으로 나타난다.
mean_values = data[['Age', 'Salary']].mean()

# 평균값 출력
print(f"Age 열의 평균값: {age_mean}")
print(f"Salary 열의 평균값: {salary_mean}")
print("Age와 Salary 열의 평균값:\n", mean_values)
```

### Chapter 5-3-2 최댓값 구하기:max() <a class="anchor" id="chapter5-3-2"></a>
1. `max()` 함수는 DataFrame의 수치형 열에 대한 최댓값을 계산하는 함수이다.
    - `max()` 함수는 기본적으로 수치형 열에 대해서만 계산하며, 비수치형 열이 혼재되어 있으면, 수치형 열에 대한 최댓값만 반환한다.
    - `max()` 함수는 `skipna=True` 옵션을 사용하여 결측값을 무시하고 최댓값을 계산할 수 있다.

```python
# example.csv 파일에서 각 열의 최댓값을 구해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 각 열의 최댓값 구하기
#   - 시리즈 형태로 반환: 열이름은 인덱스, 최댓값은 값으로 나타난다.
#   - 비수치형 데이터는 알파벳 순서로 가장 뒤에 있는 값이 최댓값으로 나타난다.
max_values = data.max()

# Salary 열의 최댓값 구하기
salary_max = data['Salary'].max()

# 최댓값 출력
print("각 열의 최댓값:\n", max_values)
print(f"Salary 열의 최댓값: {salary_max}")
```

### Chapter 5-3-3 최솟값 구하기:min() <a class="anchor" id="chapter5-3-3"></a>
1. `min()` 함수는 DataFrame의 수치형 열에 대한 최솟값을 계산하는 함수이다.
    - `min()` 함수는 기본적으로 수치형 열에 대해서만 계산하며, 비수치형 열이 혼재되어 있으면, 수치형 열에 대한 최솟값만 반환한다.
    - `min()` 함수는 `skipna=True` 옵션을 사용하여 결측값을 무시하고 최솟값을 계산할 수 있다.

```python
# example.csv 파일에서 각 열의 최솟값을 구해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 각 열의 최솟값 구하기
#   - 시리즈 형태로 반환: 열이름은 인덱스, 최솟값은 값으로 나타난다.
#   - 비수치형 데이터는 알파벳 순서로 가장 앞에 있는 값이 최솟값으로 나타난다.
min_values = data.min()

# Age 열의 최솟값 구하기
age_min = data['Age'].min()

# 최솟값 출력
print("각 열의 최솟값:\n", min_values)
print(f"Age 열의 최솟값: {age_min}")
```

### Chapter 5-3-4 데이터 모두 더하기:sum() <a class="anchor" id="chapter5-3-4"></a>
1. `sum()` 함수는 DataFrame의 수치형 열에 대한 합계를 계산하는 함수이다.
    - 문자열 데이터: 문자열을 이어 붙인다.
    - 불 데이터: True는 1로, False는 0으로 계산하여 합계를 구한다.
    - 수치형 데이터만 합산하려면 `numeric_only=True` 옵션을 사용한다.
    - 특정 열의 합계를 구하려면 `data['열이름'].sum()`과 같이 사용한다.

```python
# example.csv 파일에서 각 열의 합을 구해 보자.
# CSV 파일 경로
csv_file_path = './data/example.csv'

# CSV 파일 불러오기
data = pd.read_csv(csv_file_path)

# 수치형 데이터만 합계 구하기
#   - 시리즈 형태로 반환: 열이름은 인덱스, 합계는 값으로 나타난다.
numeric_sum = data.sum(numeric_only=True)

# 출력하기
print("수치형 데이터의 합계:\n", numeric_sum)
```
