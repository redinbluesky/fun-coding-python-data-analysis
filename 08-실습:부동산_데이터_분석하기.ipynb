## Chapter 8-2 부동산 매매가격 분석하기 <a class="anchor" id="chapter8-2"></a>
### Chapter 8-2-1 평균 매매가격 계산하기 <a class="anchor" id="chapter8-2-1"></a>

```python
# 부동산 매매 가격의 평균 계산하기
average_price = df['거래금액(만원)'].mean()
print("평균 매매 가격: ", average_price)
```

```python
# 각 시군구의 평균 매매가격 계산하기
grouped = df.groupby('시군구')
average_price = grouped['거래금액(만원)'].mean()
print("지역별 평균 매매가격:")
average_price
```

```python
# 각 단지의 최대 전용면적을 구한다.
max_ares = df.groupby('단지명')['전용면적(㎡)'].max()
print("단지별 최대 전용면적: ")
max_ares
```

```python
# 건축년도별 거래 횟수 구하기
yearly_transactions = df.groupby('건축년도')['NO'].count()
print("건축년도별 거래 횟수: ")
print(yearly_transactions)
```

## Chapter 8-3 셀프체크 <a class="anchor" id="chapter8-3"></a>

```python
# 거래금액(만원) 열의 최솟값, 최댓값, 평균, 중앙값을 계산해 출력하는 코드를 작성
df['거래금액(만원)'].min(), df['거래금액(만원)'].max(), df['거래금액(만원)'].mean(), df['거래금액(만원)'].median()
```

```python
# 시군구 열을 기준으로 각 지역의 평균 거래금액을 계산한 후, 평큔 거래금액이 높은 상위 5개 지역을 출력하는 코드 작성
average_price = df.groupby('시군구')['거래금액(만원)'].mean()
top_5 = average_price.sort_values(ascending=False).head(5)
top_5

```

```python
# 시군구 열에서 '서울특별시 강남구 신사동'의 거래 데이터만 필터링한 후, 해당 데이터에서 평균 거래금액과 최대 거래금액을 계산하는 코드 작성
df2 = df[df['시군구'] == '서울특별시 강남구 신사동']
max_price = df2['거래금액(만원)'].max()
mean_price = df2['거래금액(만원)'].mean()
print(max_price)
print(mean_price)
```
