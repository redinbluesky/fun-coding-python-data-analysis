### Chapter 3-1-1 변수 <a class="anchor" id="chapter3-1-1"></a>

### Chapter 3-1-2 조건문 <a class="anchor" id="chapter3-1-2"></a>
1. 조건문읜 특정 조건을 만족할 때만 코드를 실행하는 문장이다.
    - if, elif, else 키워드를 사용하여 조건문을 작성한다.

```python
temperature = 25

if temperature > 30:
    print("It's a hot day.")
elif temperature > 20:
    print("It's a warm day.")
else:
    print("It's a cold day.")
```

### Chapter 3-1-3 반복문 <a class="anchor" id="chapter3-1-3"></a>
1. 반복문은 특정 코드를 여러 번 반복해서 실행할 때 사용한다.
    - for 문과 while 문이 있다.

2. for 문
    - 리스트, 튜플, 문자열 등 반복 가능한 객체(iterable)의 각 요소에 대해 코드를 반복 실행한다.

```python
# for 문을 사용해 0부터 4까지 출력
#   - Range 함수는 지정한 범위의 숫자를 생성한다.
for i in range(5):
    print("Iteration:", i)
```

3. while 문
    - 특정 조건이 참인 동안 코드를 반복 실행한다.

```python
# while 문을 사용해 0부터 4까지 출력
count = 0
while count < 5:
    print("Count:", count)
    count += 1
```

### Chapter 3-1-4 함수 <a class="anchor" id="chapter3-1-4"></a>
1. 함수는 특정 작업을 수행하는 코드 블록으로, 재사용이 가능하다.
    - def 키워드를 사용하여 함수를 정의한다.

```python
# greet라는 함수를 정의한다.
def greet(name):
    return "Hello, " + name + "!"

# 함수를 호출한다.
message = greet("Bob")
print(message)
```

```python
# add 함수르 매개 변수 x와 y를 더한 값을 반환한다.
def add(x, y):
    return x + y
result = add(5, 3)
print("Sum:", result)
```

2. 내장 함수는 별도로 정의하지 않아도 바로 사용할 수 있는 함수이다.
    - 예: print(), len(), type() 등

3. 매서드(method)는 특정 객체에 속한 함수를 의미한다.
    - 예: 문자열 객체의 upper() 메서드 등

```python
text = "Hello, World!"
upper_text = text.upper()  # 문자열을 대문자로 변환하는 메서드
print("Original Text:", text)
split_text = text.split(", ")  # 문자열을 쉼표로 분리하는 메서드
print("Split Text:", split_text)
```

## Chapter 3-2 파이썬 데이터 구조 <a class="anchor" id="chapter3-2"></a>
### Chapter 3-2-1 리스트 <a class="anchor" id="chapter3-2-1"></a>
1. 리시트는 여거 값을 순서대로 하나의 변수에 저장할 수 있는, 변경 가능한 데이터 구조이다.
    - 대괄호([])를 사용하여 리스트를 정의한다.
    - 요소를 추가, 삭제, 수정할 수 있다.

```python
# "apple", "banana", "cherry"라는 세 개의 문자열 요소를 가진 fruits 리스트를 정의한다.
fruits = ["apple", "banana", "cherry"]
print("Fruits List:", fruits)
```

```python
# 첫 번재 요소 수정
fruits[0] = "orange"
print("Modified Fruits List:", fruits)
```

```python
# 새로운 요수 추가
fruits.append("grape")
print("After Appending Grape:", fruits)
```

```python
# 특정 요소 삭제 remove 메서드를 사용해 "banana" 요소를 삭제한다.
fruits.remove("banana")
print("After Removing Banana:", fruits)
```

```python
# 특정 요소 삭제  del 키워드를 사용해 인덱스 1의 요소를 삭제한다.
del fruits[1]
print("After Deleting Index 1:", fruits)
```

### Chapter 3-2-2 튜플 <a class="anchor" id="chapter3-2-2"></a>
1. 튜플은 리스트와 유사하지만, 한 번 생성되면 요소를 변경할 수 없는(immutable) 데이터 구조이다.
    - 소괄호(())를 사용하여 튜플을 정의한다.
    - 요소를 추가, 삭제, 수정할 수 없다.

2. 튜플은 변경 불가능한 특성 때문에 데이터의 무결성을 유지해야 할 때 유용하다.
    - 예: 좌표, RGB 색상 값 등

```python
# "apple", "banana", "cherry"라는 세 개의 문자열 요소를 가진 fruits 튜플을 정의한다.
fruits = ("apple", "banana", "cherry")  
print("Fruits Tuple:", fruits)
```

```python
# 인덱스를 사용해 첫 번째 요소에 접근한다.
first_fruit = fruits[0]
print("First Fruit:", first_fruit)
```

```python
# 튜플 변경 시도 (오류 발생)
fruits[0] = "orange"  
```

### Chapter 3-2-3 딕셔너리 <a class="anchor" id="chapter3-2-3"></a>
1. 딕셔너리는 키(key)와 값(value) 쌍으로 데이터를 저장하는, 변경 가능한 데이터 구조이다.
    - 중괄호({})를 사용하여 딕셔너리를 정의한다.
    - 키와 값을 콜론(:)으로 구분하고, 각 쌍은 쉼표(,)로 구분한다.
    - 키는 고유해야 하며, 값은 어떤 자료형이든 될 수 있다.

2. 딕셔너리는 키-값 쌍을 통해 데이터를 효율적으로 관리할 수 있는 강력한 도구이다.
    - 예: 학생 정보(이름, 나이, 전공 등)를 저장하는 데 유용하다.

```python
# "name", "age", "major"라는 키와 각각 "Alice", 21, "Computer Science"라는 값을 가진 student 딕셔너리를 정의한다.
student = {"name": "Alice", "age": 21, "major": "Computer Science"}
print("Student Dictionary:", student)
```

```python
# 딕셔너리 갑스이 접근 및 변경
student["age"] = 22  # 나이 갱신
print("Updated Student Dictionary:", student)
```

```python
# 딕셔너리 요소의 추가 및 삭제
student["graduation_year"] = 2024  # 졸업 연도 추가
print("After Adding Graduation Year:", student)

del student["major"]  # 전공 삭제
print("After Deleting Major:", student)
```

### Chapter 3-2-4 세트 <a class="anchor" id="chapter3-2-4"></a>
1. 세트는 중복되지 않는 고유한 요소들의 집합을 저장하는, 변경 가능한 데이터 구조이다.
    - 중괄호({})를 사용하여 세트를 정의한다.
    - 요소의 순서가 없으며, 인덱스로 접근할 수 없다.
    - 데이터의 존재 여부를 빠르게 확인할 수 있다.

2. 세트는 중복된 데이터를 제거하거나, 집합 연산(합집합, 교집합, 차집합 등)을 수행할 때 유용하다.

```python
# "apple", "banana", "cherry"라는 세 개의 문자열 요소를 가진 fruits 세트를 정의한다.
fruits = {"apple", "banana", "cherry"}
print("Fruits Set:", fruits)
```

```python
# 세트 요소의 추가 및 삭제
fruits.add("orange")  # 요소 추가
print("After Adding Orange:", fruits)

fruits.remove("banana")  # 요소 삭제
print("After Removing Banana:", fruits)
```

```python
# 이미 존재하는 요소를 추가해도 변화가 없다
fruits.add("apple")
print("After Adding Existing Apple:", fruits)
```

## Chapter 3-3 파이썬 기본 문벚과 데이터 구조 활용 <a class="anchor" id="chapter3-3"></a>
### Chapter 3-3-1 변수와 조건문: 점수에 따른 등급 출력하기 <a class="anchor" id="chapter3-3-1"></a>

```python
# input() 함수를 사용해 사용자로 부터 점수를 입력 받는다.
#   - input() 함수는 항상 문자열을 반환하므로, int() 함수를 사용해 정수로 변환한다.
#   - 변환된 점수는 score 변수에 저장한다.
score = int(input("Enter your score:"))
```

```python
# 조건문의 사용해 score 값에 따라 등급을 결정한다.
#   - 90점 이상은 'A', 80점 이상은 'B', 70점 이상은 'C', 그 외는 'F'로 등급을 부여한다.
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
```

```python
# print를 사용해 등급을 출력한다.
print("Your grade is: ", grade)
```

### Chapter 3-3-2 반복문: 1부터 10까지 숫자 출력하기 <a class="anchor" id="chapter3-3-2"></a>

```python
# for 문을 사용해 1 부터 10까지 숫자를 출력한다.
for i in range(1, 11):
    print(i)
```

```python
# while 문을 사용해 1부터 10까지 숫자를 출력한다.
i = 1
while i <= 10:
    print(i)
    i += 1
```

### Chapter 3-3-3 데이터 구조:학생 성적 관리 프로그램 <a class="anchor" id="chapter3-3-3"></a>

```python
# 딕셔너리를 활용해 학생들의 성적을 저장하고, 이를 출력, 추가, 수정, 삭제하는 프로그램을 작성한다.

# 학생 성적 딕셔너리 선언
scores = {"John": 85, "Jane":92, "Dave": 78, "Amy": 88} 

# For 문을 사용해 모든 학생의 이름과 성적을 출력한다.
for student, score in scores.items():
    print(f"{student} : {score}")
```

```python
# "Tom" 학생의 성적 추가
scores["Tome"] = 95
print("After Adding Tom:", scores)
```

```python
# Del을 통해 "Dave" 학생의 성적 삭제
del scores["Dave"]
print("After Deleting Dave:", scores)
```
