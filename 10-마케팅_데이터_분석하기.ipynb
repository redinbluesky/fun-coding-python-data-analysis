### Chapter 10-1-1 데이터 구조 파악하기 <a class="anchor" id="chapter10-1-1"></a>
1. marketing_campaign_dataset.csv 파일의 컬럼 확인
    - 인덱스: 데이터 프레임의 인덱스를 나타내며 0부터 시작한다.
    - Campaign_ID: 각 마케팅 캠페인의 고유 식별자이다.
    - Company: 마케팅 캠페인을 진행한 회사의 이름이다.
    - Campaign_Type: 마케팅 캠페인의 유형을 나타내며, 예를 들어 이메일, 소셜 미디어, TV 광고 등이 있다.
    - Target_Audience: 마케팅 캠페인의 타겟 오디언스를 나타내며, 예를 들어 남자 20-30세, 여자 30-40세 등이 있다.
    - Duration: 마케팅 캠페인의 기간을 나타내며, 예를 들어 30일, 60일 등이 있다.
    - Cnannel_Used: 마케팅 캠페인에서 사용된 채널을 나타내며, 예를 들어 이메일, 소셜 미디어, TV 광고 등이 있다.
    - Conversion_Rate: 마케팅 캠페인의 전환율을 나타내며, 예를 들어 0.05는 5%의 전환율을 의미한다.
    - Acquisition_Cost: 마케팅 캠페인에서 고객 한 명을 획득하는 데 드는 비용을 나타내며, 예를 들어 10.0은 고객 한 명을 획득하는 데 10달러가 드는 것을 의미한다.
    - ROI: 마케팅 캠페인의 투자 수익률을 나타내며, 예를 들어 0.2는 20%의 투자 수익률을 의미한다. 
        - 캠페인의 전반적인 성과를 평가하는 지표이다.
    - Location: 마케팅 캠페인이 진행된 위치를 나타내며, 예를 들어 미국, 유럽 등이 있다. 
        - 지역별 마케팅 전략을 평가하는데 사용된다.
    - Language: 마케팅 캠페인이 진행된 언어를 나타내며, 예를 들어 영어, 스페인어 등이 있다. 
        - 현지화 전략을 평가하는데 사용된다.
    - Clicks: 마케팅 캠페인에서 발생한 클릭 수를 나타낸다. 
        - 캠페인의 흥미 유발 정도를 평가한다.
    - Impressions: 마케팅 캠페인에서 발생한 노출 수를 나타낸다. 
        - 캠페인의 도달 범위를 평가하는 지표이다.
    - Engagement_Score: 마케팅 캠페인에서 발생한 참여 점수를 나타내며, 1(참여보 매우 낮음)부터 10(참여도 매우 높음) 사이에서 산출된다.
        - 클릭, 댓글 등 사용자가 캠페인에 얼마나 적극적으로 참여했는지를 평가하는 지표이다.
    - Customer_Segment: 마케팅 캠페인의 고객 세그먼트를 나타내며, 예를 들어 고소득층, 중산층 등이 있다. 
        - 특정 고객 세그먼트에 대한 캠페인 성과를 평가하는데 사용된다.
    - Date: 마케팅 캠페인이 진행된 날짜를 나타낸다. 
        - 시간에 따른 캠페인 성과를 평가하는데 사용된다.

## Chapter 10-2 데이터 분석하기 <a class="anchor" id="chapter10-2"></a>
### Chapter 10-2-1 데이터 전처리하기 <a class="anchor" id="chapter10-2-1"></a>

```python
# 고객 획득비용(Acquisition_Cost) 컬럼 값의 5 개행 출력
df['Acquisition_Cost'].head()
```

```python
# 고객 획득비용(Acquisition_Cost) 컬럼의 특수기호 제거 후 float 형으로 변환
df['Acquisition_Cost'] = df['Acquisition_Cost'].str.replace('$','').str.replace(',','').astype(float);
```

```python
# 총 고객 획득비용 계산
total_acquisition_cost = df['Acquisition_Cost'].sum();
total_acquisition_cost
```

### Chapter 10-2-2 마케팅 채널별 고객 획득 비용 분석하기 <a class="anchor" id="chapter10-2-2"></a>

```python
# Channel_Used 컬럼으로 그룹화 후 합계 금액을 구한다.
channel_acquisition_cost = df.groupby('Channel_Used')['Acquisition_Cost'].sum()
print("채널별 고객 획득 비용:")
print(channel_acquisition_cost)
```

```python
# CSV 파일로 저장
channel_acquisition_cost.to_csv('./data/channel_acquisition_cost.csv')
```

### Chapter 10-2-3 마케팅 채널별 평균 구매 전환율 분석하기 <a class="anchor" id="chapter10-2-3"></a>

```python
# Channel_Used 그룹화 후 Conversion_Rate열의 평균을 계산한다.
channel_conversion_rate = df.groupby('Channel_Used')['Conversion_Rate'].mean()
print("채널별 평균 구매 전환율: ")
print(channel_conversion_rate)
```

```python
# CSV 파일로 저장
channel_conversion_rate.to_csv('./data/channel_conversion_rate.csv')
```

### Chapter 10-2-4 타깃 고객 그룹별 평균 구매 전황율 분석하기 <a class="anchor" id="chapter10-2-4"></a>

```python
# Target_Audience 열을 기준으로 그룹화 후 Conversion_Rate 열의 평균을 구한다.
audience_conversion_rate = df.groupby('Target_Audience')['Conversion_Rate'].mean()
print("타깃 고객 그룹별 평균 구매 전환율: ")
print(audience_conversion_rate)
```

```python
# CSV 파일로 저장
audience_conversion_rate.to_csv('./data/audience_conversion_rate.csv')
```

### Chapter 10-2-5 캠페인 기간별 구매 전환율 분석하기 <a class="anchor" id="chapter10-2-5"></a>

```python
# Duration열을 기준으로 그룹화 후 Conversion_Rate 열의 평균 계산
duration_conversion_rate = df.groupby('Duration')['Conversion_Rate'].mean()
print("캠페인 기간별 평균 구매 전환율: ")
print(duration_conversion_rate)
```

```python
# CSV 파일로 저장
duration_conversion_rate.to_csv('./data/duration_conversion_rate.csv')
```

### Chapter 10-2-6 지역별 고객 획득 비용 분석하기 <a class="anchor" id="chapter10-2-6"></a>

```python
# Location 열을 기준으로 그룹화 후 Acauisition_Cost열의 합 구하기
location_acquisition_cost = df.groupby('Location')['Acquisition_Cost'].sum()
print("지역별 고객 획득 비용: ")
print(location_acquisition_cost)
```
